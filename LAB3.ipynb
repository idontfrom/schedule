```python
import numpy as np
import networkx as nx
import random
import matplotlib.pyplot as plt

class Node:
    def __init__(self, number):
        self.predecessors = []
        self.successors = []
        self.number = number
        self.dur = 0

def GraphGenerator(n : int):
    time_list = [1,2]
    edge_rand = [0,1]
    nodes : list[Node] = []
    graph_matrix=np.zeros((n,n), dtype=int)
    graph = nx.DiGraph()

    for i in range(n):
        node = Node(i)

        node.dur = random.choice(time_list)


        nodes.append(node)
    #Протягивать ребра будем только к узлам с большим номером
    for i in range(n):
        for j in range(i+1,n):
            coin_flip = random.choice(edge_rand)
            if coin_flip == 1:
                nodes[i].successors.append(i)
                nodes[j].successors.append(j)
                graph_matrix[i][j] = 1
                graph_matrix[j][i] = -1
                graph.add_edge
    #Визуализируем граф
    for i in range(n):
        graph.add_node(i)
        graph.nodes[i]['time'] = nodes[i].dur

    for i in range(n):
        for j in range(n):
            if (graph_matrix[i][j] == 1):
                graph.add_edge(i,j)
    #nx.draw_networkx(graph, position, node_color=color_map)
    node_labels = nx.get_node_attributes(graph, 'time')
    position = nx.random_layout(graph)
    attr_pos = {}
    for key in position:
        attr_pos.update({key : (position[key][0], position[key][1] - 10)})
    nx.draw(graph, position, with_labels=True, font_weight='bold', font_color='whitesmoke')
    nx.draw_networkx_labels(graph, attr_pos, labels=node_labels, font_size=10)
    plt.show()
    return nodes, graph_matrix

GraphGenerator(5)
```
